```python
import requests
from bs4 import BeautifulSoup
import IPython.display as display
```

```python
pesquisa = str(input('Digite o termo a ser pesquisado\n'))
```

```python
r = requests.get("https://images.search.yahoo.com/search/images", \
                     params={'fr2':'sb-top-images.search','p':f'{pesquisa.replace(" ","+")}'})
```

```python
soup = BeautifulSoup(r.text,'html.parser')
```

```python
#soup.find("img", {"alt": "Image result for fiat punto"})
soup.find_all("img")[0]['data-src']
```

```python
display.Image(requests.get(soup.find_all("img")[0]['data-src']).content)
```

### Abaixo irá capturar fotos de vários modelos

```python
#carros = ['fiat palio','ford fiesta titanium','ford ka+','ford ecosport 2019','chevrolet corsa classic']
carros = ['fiat palio']

for carro in carros:
    r = requests.get("https://images.search.yahoo.com/search/images", \
                     params={'fr2':'sb-top-images.search','p':f'{carro.replace(" ","+")}'})

    soup = BeautifulSoup(r.text,'html.parser') 
    print('busca = '+carro)
    soup.find_all("img")[0]['data-src']
display.Image(requests.get(soup.find_all("img")[0]['data-src']).content)
```
